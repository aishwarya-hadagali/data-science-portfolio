```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

cancer = load_breast_cancer()
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
df['diagnosis'] = cancer.target  # 0 = malignant, 1 = benign

print(df.shape)
print(df.head())
sns.histplot(data=df, x='mean radius', hue='diagnosis',  kde=True)
plt.title('Distribution of Mean Radius')
plt.show()

plt.figure(figsize=(15, 15))
numeric_df = df.select_dtypes(include=['float64'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

sns.barplot(data=df, x='diagnosis', y='mean texture', hue='diagnosis')
plt.title('Mean Texture by Diagnosis')
plt.xlabel('Diagnosis')
plt.ylabel('Mean Texture')
plt.show()
cols = ['mean radius', 'mean smoothness', 'worst concavity', 'diagnosis']
sns.pairplot(data=df[cols], hue='diagnosis')
plt.show()

sns.boxplot(data=df, x='diagnosis', y='worst texture', palette='husl', hue='diagnosis')
plt.title('Worst Texture Distribution by Diagnosis')
plt.xlabel('Diagnosis')
plt.ylabel('Worst Texture')
plt.show()


```

```python
df.to_csv('breast_cancer_data.csv', index=False)
print('DataFrame saved to breast_cancer_data.csv')
```
