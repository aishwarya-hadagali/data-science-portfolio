```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('penguins')
df.head()
df.info()
df.shape
df.isnull().sum()
df.duplicated().sum()
df.describe()

df['bill_length_mm'].fillna(df['bill_length_mm'].median(), inplace=True)
df['bill_depth_mm'].fillna(df['bill_depth_mm'].mean(), inplace=True)
df['flipper_length_mm'].fillna(df['flipper_length_mm'].median(), inplace=True)
df['body_mass_g'].fillna(df['body_mass_g'].median(), inplace=True)
df['sex'].fillna(df['sex'].mode()[0], inplace=True)
df.isnull().sum()

df.describe()
df.describe(include='object')
df.describe()

#univariate Analysis
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle("Penguins Dataset — EDA", fontsize=16, fontweight='bold')

# Plot 1 — top left (axes[0,0])
sns.histplot(data=df, x='body_mass_g',ax=axes[0, 0])

# Plot 2 — top right (axes[0,1])
sns.countplot(data=df, x='species',ax=axes[0,1])

# Plot 3 — bottom left (axes[1,0])
sns.countplot(data=df,x='island',ax=axes[1,0])

# Plot 4 — bottom right (axes[1,1])
sns.boxplot(x=df['flipper_length_mm'], ax=axes[1,1])

axes[0,0].set_title('Distribution of Body Mass (g)')
axes[0,1].set_title('Count of Species')
axes[1,0].set_title('Count of Islands')
axes[1,1].set_title('Box Plot of Flipper Length (mm)')

plt.tight_layout()
plt.show()

#Bivariate Analysis
fig, ax = plt.subplots(figsize=(8, 5))
sns.boxplot(data=df, x='species', y='body_mass_g', ax=ax)
ax.set_title('Body Mass by Species')
plt.show()
```
