```python
import requests
import pandas as pd
import numpy as np

BASE_URL = "http://4.186.26.27:8787"

response = requests.get(f"{BASE_URL}/ecommerce?dataset=youtube")
df = pd.DataFrame(response.json())
df.shape
df.head()
df.isnull().sum()
df.replace('null', np.nan, inplace=True)
df.dropna(inplace=True)
df.isnull().sum()
df.shape
df.drop(columns=['Abbreviation', 'Latitude', 'Longitude'], inplace=True)
df.columns
df['engagement_rate'] = (df['subscribers_for_last_30_days'] / df['subscribers']) * 100
df['views_per_upload'] = (df['video views'] / df['uploads'])
print(df.shape)
print(df.isnull().sum())
print(df.head())





```
