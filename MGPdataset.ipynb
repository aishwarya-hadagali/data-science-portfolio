```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('mpg')
df.shape
df.info()
df.isnull().sum()
df.duplicated().sum()
df.describe()
df['horsepower'] = df['horsepower'].fillna(df['horsepower'].median())
df.isnull().sum()

fig,axes = plt.subplots(2,2,figsize=(15,10))
fig.suptitle("MPG dataset 🚗", fontweight='bold')
sns.histplot(df['mpg'],ax=axes[0,0])
sns.histplot(df['cylinders'],ax=axes[0,1])
sns.histplot(df['displacement'],ax=axes[1,0])
sns.histplot(df['weight'],ax=axes[1,1])

axes[0,0].set_title('mpg')
axes[0,1].set_title('cylinders')
axes[1,0].set_title('displacement')
axes[1,1].set_title('weight')
plt.show()


fig,axes = plt.subplots(2,2,figsize=(15,10))
fig.suptitle("MPG dataset 🚗", fontweight='bold')
sns.scatterplot(x='mpg',y='weight',data=df,ax=axes[0,0])
sns.scatterplot(x='mpg',y='displacement',data=df,ax=axes[0,1])
sns.scatterplot(x='mpg',y='horsepower',data=df,ax=axes[1,0])
sns.scatterplot(x='mpg',y='acceleration',data=df,ax=axes[1,1])

axes[0,0].set_title('mpg vs weight')
axes[0,1].set_title('mpg vs displacement')
axes[1,0].set_title('mpg vs horsepower')
axes[1,1].set_title('mpg vs acceleration')
plt.show()



sns.pairplot(data=df,hue='origin')
plt.show()
```
