```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('diamonds')
df.head()
df.shape
df.info()
df.dtypes

#clean
df.isnull().sum()
df.duplicated().sum()
df.drop_duplicates(inplace=True)
df.shape
df.isnull().sum()
df.describe()
df[df['x'] == 0].shape[0]   # count of rows where x is 0
df[df['y'] == 0].shape[0]   # count of rows where y is 0
df[df['z'] == 0].shape[0]   # count of rows where z is 0

df = df[(df['x'] != 0) & (df['y'] != 0) & (df['z'] != 0)]
df.shape
df.isnull().sum()

df.describe(include='category')

#univariate
fig,axes = plt.subplots(2,2,figsize=(15,10))
sns.histplot(data=df,x='price',ax=axes[0,0])
sns.histplot(data=df,x='carat',ax=axes[0,1])
sns.countplot(data=df,x='cut',ax=axes[1,0])
sns.countplot(data=df,x='color',ax=axes[1,1])

axes[0,0].set_title('price')
axes[0,1].set_title('carat')
axes[1,0].set_title('cut')
axes[1,1].set_title('color')

#bivariate
fig,axes = plt.subplots(2,2,figsize=(15,10))
sns.scatterplot(data=df,x='carat',y='price',ax=axes[0,0])
sns.boxplot(data=df,x='cut',y='price',ax=axes[0,1])
sns.boxplot(data=df,x='color',y='price',ax=axes[1,0])
sns.boxplot(data=df,x='clarity',y='price',ax=axes[1,1])

axes[0,0].set_title('carat vs price')
axes[0,1].set_title('cut vs price')
axes[1,0].set_title('color vs price')
axes[1,1].set_title('clarity vs price')
```
